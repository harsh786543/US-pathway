```python
from scipy import stats

def detect_outliers(data, threshold=3):
    z_scores = np.abs(stats.zscore(data))
    return z_scores > threshold


numeric_columns = df.select_dtypes(include=[np.number]).columns

outlier_counts = {}

for column in numeric_columns:
    outliers = detect_outliers(df[column].dropna())
    outlier_counts[column] = outliers.sum()

for column, count in outlier_counts.items():
    print(f"{column}: {count}")

print(f"Total number of rows: {len(df)}")

plt.figure(figsize=(15, 20))
i = 1
for column, count in outlier_counts.items():
    if count > 0:
        outliers = df[detect_outliers(df[column].dropna())]
        plt.subplot(len(outlier_counts) // 2 + 1, 2, i)
        plt.scatter(df.index, df[column], label='Data')
        plt.scatter(outliers.index, outliers[column], color='red', label='Outliers')
        plt.title(f'{column.capitalize()} Outliers')
        plt.xlabel('Index')
        plt.ylabel(column.capitalize())
        plt.legend()
        i += 1

plt.tight_layout()
plt.show()
```

```python
df['total_spent'] = pd.to_numeric(df['total_spent'].replace('[\$,]', '', regex=True), errors='coerce')

def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df

columns = ['total_spent', 'impressions', 'clicks', 'leads']


fig, axs = plt.subplots(len(columns), 2, figsize=(20, 5 * len(columns)))

for i, column in enumerate(columns):
    sns.boxplot(x=df[column], ax=axs[i, 0])
    axs[i, 0].set_title(f'{column} - Before handling outliers')
    
    df = handle_outliers(df, column)
    sns.boxplot(x=df[column], ax=axs[i, 1])
    axs[i, 1].set_title(f'{column} - After handling outliers')

plt.tight_layout()
plt.show()
fig, axs = plt.subplots(2, 2, figsize=(20, 15))
axs = axs.ravel()

for i, column in enumerate(columns):
    sns.histplot(df[column], kde=True, ax=axs[i])
    axs[i].set_title(f'Distribution of {column} after handling outliers')
    axs[i].set_xlabel(column)

plt.tight_layout()
plt.show()

print(df[columns].describe())
```
